```python
BYU_UW_root = r"G:\.shortcut-targets-by-id\1k1B8zPb3T8H7y6x0irFZnzzmfQPHMRPx\Illimited Lab Projects\Research Projects\Spiders\BYU-UW"
color_vec = plt.rcParams['axes.prop_cycle'].by_key()['color']

# --- New Configuration for Plot Style ---
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    'font.size': 20,       # General font size increased
    'axes.labelsize': 22,  # Axis label font size
    'xtick.labelsize': 18, # X-tick labels font size
    'ytick.labelsize': 18, # Y-tick labels font size
    'legend.fontsize': 16, # Legend font size
})
# Create the figure with a larger width
fig, ax = plt.subplots(figsize=(10, 6))
# --- End of New Configuration ---

ax2 = ax.twinx()
ax.set_xlabel('Maximum stress (MPa)')
ax2.set_ylabel('Strain (-)', color=color_vec[1])
ax.set_ylabel('Relative error (\%)', color=color_vec[0])
ax.set_xlim(0, stress_data[-1])
# ax2.set_ylim(0, strain_data[-1])
error_lim = 6
ax.set_ylim(0, error_lim)
ax2.set_ylim(0, error_lim/10)
ax2.plot(stress_data, strain_data,'-',alpha = 0.5, lw = 1, label='Mean Stress-Strain Study', color = color_vec[1])

calibration_results_files = glob.glob(os.path.join(BYU_UW_root, 'calibration_results','new specs', '*_cr.npy'))
tol_mm = 0.1
for file in calibration_results_files:
    if not "fstudy" in file:
        continue
    data = np.load(file, allow_pickle=True).item()
    model_name_i = data['model_name']
    real_vertices = data['real_vertices']
    vertices      = data['vertices']
    print(model_name_i)
    net2 = Network_custom.load_network(os.path.join(BYU_UW_root, 'networks', model_name_i + '.pkl'))

    f = net2.f
    sigma = net2.f/net2.A[0]
    sigma_unit_edges = sigma[unit_edges]
    Rmin = np.min(net2.R[unit_edges])
    simga_max = np.max(sigma)
    error_edge = []
    tol_error_edge = []
    
    for edge_i, edge in enumerate(net2.edges):
        coor0, coor1 = real_vertices[edge[0]], real_vertices[edge[1]]
        l1_meas = np.linalg.norm(coor0[:2] - coor1[:2])

        coor2, coor3 = vertices[edge[0]], vertices[edge[1]]
        l1_ref = np.linalg.norm(coor2[:2] - coor3[:2])

        error_edge.append(np.abs(l1_meas - l1_ref) / l1_ref)

        # Relative tolerance in percentage
        rel_tol = tol_mm / l1_ref
        tol_error_edge.append(rel_tol)
    edge_error = np.array(error_edge) * 100
    edge_error_mean = np.mean(edge_error[unit_edges])
    edge_tol   = np.array(tol_error_edge) * 100
    edge_tol_mean = np.mean(edge_tol[unit_edges])
    
    # Corrected lower error calculation
    lower_err = edge_error_mean - np.maximum(0.01, edge_error_mean - edge_tol_mean)
    upper_err = edge_tol_mean
    yerr_asymmetric = np.array([lower_err, upper_err]).reshape(2,1)
    ax.plot(simga_max*np.ones_like(edge_error[unit_edges]), edge_error[unit_edges], '.', color = 'b', alpha = 0.3, zorder=0, label = "Individual edges")
    ax.errorbar(
        simga_max,
        edge_error_mean,
        yerr=yerr_asymmetric,
        fmt='o',
        capsize=3,
        # label=model_name_i,
        color = color_vec[0],
        zorder=1,
        label = "Mean edge error"
    )
    # ax2.plot(simga_max, 1/Rmin,'k*',  label = 'Max curvature')

    strain_max = strain_from_stress(simga_max)
    ax2.plot(simga_max, strain_max, '*', color=color_vec[1], label = "Max strain in specimen")

handles, labels = ax.get_legend_handles_labels()

# Create a dictionary to map unique labels to their handles
unique = dict(zip(labels, handles))

handles2, labels2 = ax2.get_legend_handles_labels()

# Create a dictionary to map unique labels to their handles
unique2 = dict(zip(labels2, handles2))

# Pass the unique handles and labels to the legend function
legend = ax.legend(unique.values(), unique.keys())
legend2 = ax2.legend(unique2.values(), unique2.keys())
# Get the text objects from the legend and set their color
for text in legend.get_texts():
    text.set_color(color_vec[0])
for text in legend2.get_texts():
    text.set_color(color_vec[1])

ax.grid(True) 
plt.tight_layout()
pos_ax1 = ax.get_position()
plt.show()
fig.savefig(os.path.join(BYU_UW_root,'images', "stress_study_validation.png"), dpi=300)
fig.savefig(os.path.join(BYU_UW_root,'images', "stress_study_validation.svg"))
```
